```python
import os
from transformers import pipeline
import pickle

# Force progress bars to show
os.environ["HF_HUB_DISABLE_PROGRESS_BARS"] = "0"
```

```python
# Load the Model
print("Loading model... Check the bar below.")
model_id = "superb/wav2vec2-base-superb-er"
classifier = pipeline("audio-classification", model=model_id)
print("Model Loaded Successfully!")
```

```python
# Save the model components properly
classifier.model.save_pretrained("./my_emotion_model")
classifier.feature_extractor.save_pretrained("./my_emotion_model")

# Zip it for your local PC
!zip -r my_emotion_model.zip ./my_emotion_model

from google.colab import files
files.download("my_emotion_model.zip")
```
